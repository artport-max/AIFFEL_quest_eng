```python
import random

class Account:
    BANK_NAME = "SC은행"
    HOLDER_NAME = "artport"

    def __init__(self, account_type, balance):
        self.bank = Account.BANK_NAME
        self.holder = Account.HOLDER_NAME
        self.account_type = account_type
        self.balance = balance

        self.account_number = self._generate_account_number()
        self.deposit_count = 0
        self.deposit_log = []
        self.withdraw_log = []

    def _generate_account_number(self):
        return f"{random.randint(100,999)}-{random.randint(10,99)}-{random.randint(100000,999999)}"
```

```python
    # 입금
    def deposit(self, amount):
        if amount <= 0:
            return

        self.balance += amount
        self.deposit_count += 1
        self.deposit_log.append(amount)

        # 입금 5회마다 잔고의 1% 이자 지급
        if self.deposit_count % 5 == 0:
            interest = int(self.balance * 0.01)
            self.balance += interest
```

```python
    # 출금
    def withdraw(self, amount):
        if amount <= 0 or amount > self.balance:
            return

        self.balance -= amount
        self.withdraw_log.append(amount)
```

```python
    # 계좌 정보 출력
    def display_info(self):
        print("-" * 40)
        print(f"은행명: {self.bank}")
        print(f"예금주: {self.holder}")
        print(f"계좌번호: {self.account_number}")
        print(f"계좌종류: {self.account_type}")
        print(f"잔고: {self.balance:,}원")
```

```python
    # 입금 내역
    def deposit_history(self):
        print(f"[{self.holder}] 입금 내역")
        for i, amount in enumerate(self.deposit_log, 1):
            print(f"{i}회차: {amount:,}원")
```

```python
    # 출금 내역
    def withdraw_history(self):
        print(f"[{self.holder}] 출금 내역")
        for i, amount in enumerate(self.withdraw_log, 1):
            print(f"{i}회차: {amount:,}원")
```
