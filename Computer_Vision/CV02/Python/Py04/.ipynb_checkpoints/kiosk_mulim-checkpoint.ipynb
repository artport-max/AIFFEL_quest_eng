```python
# 평행사변형 메서드
def par(self):
    # 가로(밑변)와 높이를 입력받습니다.
    width, height = map(int, input('밑변, 높이를 입력하세요. 예시: 10,20\n >>>').split(','))
    # 공식: 밑변 * 높이
    area = width * height
    result = f"평행사변형의 넓이는 : {area}"
    return result

# 사다리꼴 메서드
def trape(self):
    # 밑변, 윗변, 높이를 입력받습니다.
    bottom, top, height = map(int, input('밑변, 윗변, 높이를 입력하세요. 예시: 10,5,8\n >>>').split(','))
    # 공식: (밑변 + 윗변) * 높이 / 2
    area = (bottom + top) * height / 2
    result = f"사다리꼴의 넓이는 : {area}"
    return result
```

```python


class Kiosk:
    def __init__(self):
       self.menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
       self.price = [2000, 3000, 3000, 2500, 2500, 3000]

    # 메뉴 출력 메서드
    def menu_print(self):
        for i in range(len(self.menu)):
            print(i + 1, self.menu[i], ' : ', self.price[i], '원')

    # 주문 메서드
    def menu_select(self):
        print()
        n = 0
        while n < 1 or len(self.menu) < n:
            n = int(input("음료 번호를 입력하세요 : "))
            if 1 <= n and n <= len(self.menu):
                self.price_sum = self.price[n-1]
                self.order_menu = self.menu[n-1]
            else:
                print("없는 메뉴입니다. 다시 주문해 주세요.")
        
        t = 0
        while t != 1 and t != 2:
            t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하시오 : "))
            if t == 1:
                self.temp = "HOT"
            elif t == 2:
                self.temp = "ICE"
            else:    
                print("1과 2 중 하나를 입력하세요.\n")

    # 지불
    def pay(self):
        payment = input("지불 수단을 선택하세요 (현금: 'cash' 또는 1 / 카드: 'card' 또는 2) : ")
        # 현금을 선택한 경우 (문자열 '1' 또는 'cash')
        if payment == 'cash' or payment == '1':
            print('직원을 호출하겠습니다.')

        # 카드를 선택한 경우 (문자열 '2' 또는 'card')
        elif payment == 'card' or payment == '2':
            print('IC칩 방향에 맞게 카드를 꽂아주세요.')
            
        else:
            print('잘못된 입력입니다. 다시 시도해 주세요.')

    # 주문서 출력 
    def table(self):
        # 외곽선을 그려서 주문표처럼 보이게 구성합니다.
        print('⟝' + '-' * 30 + '⟞')
        
        # 주문 내역 출력 (온도와 메뉴명을 합쳐서 출력)
        order_detail = f"{self.temp} {self.order_menu}"
        print(f'| {order_detail:<20} | {self.price_sum:>5}원 |')
        
        print('⟝' + '-' * 30 + '⟞')
        
        # 최종 합계 금액 출력
        print(f'| {"합계":<20} | {self.price_sum:>5}원 |')
        print('⟝' + '-' * 30 + '⟞')

        
```
