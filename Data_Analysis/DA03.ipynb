```python
import pandas as pd
train = pd.read_csv('train.csv')
print
```

```python
# 1. 값을 실제로 저장합니다.
train['CryoSleep'] = train['CryoSleep'].fillna(False)

print(train['CryoSleep'].isnull().sum())
```

```python
# Age 컬럼의 평균값을 구합니다.
average_age = train['Age'].mean()

print("승객 평균 나이:", average_age)
```

```python
# Age의 결측치를 위에서 구한 평균값(average_age)으로 채웁니다.
train['Age'] = train['Age'].fillna(average_age)

print("Age 결측치 개수:", train['Age'].isnull().sum())
```

```python
# RoomService의 결측치를 0으로 채웁니다.
train['RoomService'] = train['RoomService'].fillna(0)

print("RoomService 결측치 개수:", train['RoomService'].isnull().sum())
```

```python
# FoodCourt의 결측치를 0으로 채웁니다.
train['FoodCourt'] = train['FoodCourt'].fillna(0)

print("FoodCourt 결측치 개수:", train['FoodCourt'].isnull().sum())
```

```python
# ShoppingMall의 결측치를 0으로 채웁니다.
train['ShoppingMall'] = train['ShoppingMall'].fillna(0)

print("ShoppingMall 결측치 개수:", train['ShoppingMall'].isnull().sum())
```

```python
# Spa의 결측치를 0으로 채웁니다.
train['Spa'] = train['Spa'].fillna(0)

print("Spa 결측치 개수:", train['Spa'].isnull().sum())
```

```python
# Cabin과 Name 컬럼의 실제 데이터 형태 확인
print(train[['Cabin', 'Name']].head())
```

```python
# Cabin 데이터를 '/' 기준으로 나누고, 그 중 첫 번째(0번) 값인 데크만 가져옵니다.
# 결측치(NaN)가 있으면 에러가 날 수 있으므로, str을 이용해 안전하게 처리합니다.
train['Deck'] = train['Cabin'].str.split('/').str[0]

print(train[['Cabin', 'Deck']].head())
```

```python
train['Side'] = train['Cabin'].str.split('/').str[2] 
print(train[['Cabin', 'Deck', 'Side']].head())
```

- 글자로 된 카테고리를 숫자로 변환하기 # 컴퓨터가 쉽게 읽을 수 있게

```python
# HomePlanet 컬럼을 0과 1의 숫자로 변환합니다.
train = pd.get_dummies(train, columns=['HomePlanet'])

print(train.columns)
```

```python
# Side 컬럼을 원-핫 인코딩으로 변환하기
train = pd.get_dummies(train, columns=['Side'])

print(train.columns)
```

```python
# VIP 컬럼의 값별 개수 확인
print(train['VIP'].value_counts())
```

```python
# VIP 컬럼의 결측치를 False로 채웁니다.
train['VIP'] = train['VIP'].fillna(False)

print("VIP 결측치 개수:", train['VIP'].isnull().sum())
```

```python
# Destination 컬럼의 값별 개수(분포)를 확인합니다.
print(train['Destination'].value_counts())
```

```python
# Destination의 결측치를 가장 빈도가 높은 'TRAPPIST-1e'로 채웁니다.
train['Destination'] = train['Destination'].fillna('TRAPPIST-1e')

print("Destination 결측치 개수:", train['Destination'].isnull().sum())
```

```python
# Destination 컬럼을 원-핫 인코딩으로 변환합니다.
train = pd.get_dummies(train, columns=['Destination'])

print(train.columns)
```

```python
# 남은 글자 데이터(HomePlanet, Deck, Side)를 한 번에 숫자로 바꿉니다.
train = pd.get_dummies(train, columns=['HomePlanet', 'Deck', 'Side'])

# 변환된 컬럼들 확인
print(train.columns)
```

```python
# 1. Cabin에서 Deck과 Side를 다시 추출합니다.
train['Deck'] = train['Cabin'].str.split('/').str[0]
train['Side'] = train['Cabin'].str.split('/').str[2]

# 2. 잘 만들어졌는지 확인 (Deck과 Side가 목록에 있어야 합니다)
print(train.columns)
```

```python
# 남은 글자 데이터(HomePlanet, Deck, Side)를 한 번에 숫자로 바꿉니다.
train = pd.get_dummies(train, columns=['HomePlanet', 'Deck', 'Side'])

# 변환된 컬럼들 확인
print(train.columns)
```

```python
# 'Name' 컬럼에서 성(Last Name)만 추출하여 'LastName' 컬럼을 만듭니다.
# 'Maham Ofracculy' -> 'Ofracculy'
train['LastName'] = train['Name'].str.split().str[-1]

# 잘 추출되었는지 확인해봅시다.
print(train[['Name', 'LastName']].head())
```

```python
# 성씨(LastName)를 기준으로 그룹을 묶어, 각 성씨가 몇 번 나오는지 계산합니다.
train['FamilySize'] = train.groupby('LastName')['LastName'].transform('count')

# 혼자 온 승객(결측치 등으로 인해 계산이 안 된 경우)은 최소 1명으로 채워줍니다.
train['FamilySize'] = train['FamilySize'].fillna(1)

# 성씨와 가족 규모가 잘 매칭되었는지 확인해봅시다.
print(train[['LastName', 'FamilySize']].head(10))
```

```python
# 분석에 사용하지 않을 글자 컬럼들을 삭제합니다.
train_final = train.drop(['PassengerId', 'Cabin', 'Name', 'LastName'], axis=1)

# 최종적으로 남은 컬럼들과 데이터 형태를 확인합니다.
print(train_final.info())
```

- 위 object(1)에 대한 검색은 수면중인지, 소비없이 이동 중인지, 등 결측치를 다시 조정

```python
# 돈을 한 푼도 안 쓴(0원) 사람들의 CryoSleep 상태 확인
zero_spending = train[
    (train['RoomService'] == 0) & 
    (train['FoodCourt'] == 0) & 
    (train['ShoppingMall'] == 0) & 
    (train['Spa'] == 0) & 
    (train['VRDeck'] == 0)
]

print(zero_spending['CryoSleep'].value_counts())
```

```python
# 모든 지출 항목을 더해서 TotalSpending을 만듭니다.
# 빈칸(NaN)은 0으로 간주하고 더합니다.
train['TotalSpending'] = train[['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']].sum(axis=1)

# 이제 이 TotalSpending을 보고 CryoSleep의 결측치를 채울 수 있습니다.
# 만약 TotalSpending이 0보다 크다면, CryoSleep은 무엇이어야 할까요?
```

```python
# 1. 지출이 0원보다 큰 경우 (깨어 있음)
train.loc[(train['CryoSleep'].isnull()) & (train['TotalSpending'] > 0), 'CryoSleep'] = False

# 2. 지출이 0원인 경우 (잠자는 중)
train.loc[(train['CryoSleep'].isnull()) & (train['TotalSpending'] == 0), 'CryoSleep'] = True

# 잘 채워졌는지 확인 (0이 나와야 성공!)
print("CryoSleep 남은 결측치:", train['CryoSleep'].isnull().sum())
```

```python
# CryoSleep 컬럼을 bool(참/거짓) 타입으로 변환합니다.
train_final['CryoSleep'] = train_final['CryoSleep'].astype(bool)

# 최종 데이터 상태 확인
train_final.info()
```
